```python
#多个图的绘制方法（plt.subplot）
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='STHeiti' 
#要绘图的数据
month=['1','2','3','4']
sales=[100,150,80,130]
f1=plt.subplot(2,2,1) #两行两列（第一个图）
f1.plot(month,sales)
f2=plt.subplot(2,2,2) #两行两列（第二个图）
f2.bar(month,sales)
f3=plt.subplot(2,2,3) #两行两列（第三个图）
f3.scatter(month,sales)
f4=plt.subplot(2,2,4) #两行两列（第四个图）
f4.barh(month,sales)
```
