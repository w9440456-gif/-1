```python
"""
折线图 plot:趋势随时间变化
条形图 bar:类别之间对比
饼图 pie:整体比例
散点图 scatter:两变量之间关系
箱线图 boxplot:数据分布、异常
多个图表
组合图
"""
```

```python
#绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams #jupyter不带字体(1月···)，要设置
rcParams['font.family']='STHeiti' #mac的   window:SimHei
#创建图表
plt.figure(figsize=(10,5)) #图表纸张的大小
#要绘图的数据
month=['1月','2月','3月','4月']
sales=[100,150,80,130]
#绘制折线图
plt.plot(month,sales,
        label='产品A',     #label='产品A'：真正的图例
        color='orange',   
        linewidth=2,       #线的宽度
        linestyle='--',   
        marker='o')        #标记点
#添加标题
plt.title('2025年销售趋势',color='red',fontsize=20)
#添加坐标轴的标签
plt.xlabel('月份',fontsize=10)
plt.ylabel('销售额（万元）',fontsize=10)
#添加图例的位置
plt.legend(loc='upper left')
#添加网格线
plt.grid(True,alpha=0.2,color='blue',linestyle='--') #x,y方向都有网格线,alpha是透明度,linestyle为网格线风式，-：实线，--：虚线
#plt.grid(axis='y') #y方向有网格线
#设置刻度字体风格
plt.xticks(rotation=300,fontsize=12) #rotation表示刻度字体旋转的方向
plt.yticks(rotation=0,fontsize=12)
#设置y轴的范围
plt.ylim(0,160)
#每个数据点显示数值
for x,y in zip(month,sales):
       #将打包遍历出的放在相对应地x,y轴上，水平剧中，线的上面
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10)
#自动优化排版
plt.tight_layout()
#显示图表
plt.show()
```
