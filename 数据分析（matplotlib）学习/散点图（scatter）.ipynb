```python
#绘制散点图
import matplotlib.pyplot as plt
from matplotlib import rcParams #jupyter不带字体(1月···)，要设置
rcParams['font.family']='STHeiti' #mac的   window:SimHei
#创建图表
plt.figure(figsize=(10,5)) #图表纸张的大小
#要绘图的数据
scores=[50,55,60,65,70,80,85]
hours=[1,2,3,4,5,6,7]
#绘制散点图
plt.scatter(hours,scores)
#自动优化排版
plt.tight_layout()
#显示图表
plt.show()
```

```python
#绘制一个实例的散点图
import matplotlib.pyplot as plt
import random
from matplotlib import rcParams #jupyter不带字体(1月···)，要设置
rcParams['font.family']='STHeiti' #mac的   window:SimHei
#创建图表
plt.figure(figsize=(10,8)) #图表纸张的大小
#要绘图的数据
x = []
y = []
for i in range(1000): #生成1000次
    tmp=random.uniform(0,10)       #0-10之间的随机小数：   randint为整数,uniform为小数
    x.append(tmp)
    y.append(2*tmp+random.gauss(0,2))   #gauss高斯分布生成的一个数
#绘制散点图
plt.scatter(x,y,
           color='blue',
           alpha=0.5,
           s=20,     #圆点大小
           label='数据')       
#添加标题 
plt.title('x变量与y变量的关系',color='red',fontsize=20)
#添加坐标轴的标签
plt.xlabel('x自变量',fontsize=10)
plt.ylabel('y因变量',fontsize=10)
#添加图例的位置
plt.legend(loc='upper left')
#添加网格线
plt.grid(True,alpha=0.2,color='blue',linestyle='--') #x,y方向都有网格线,alpha是透明度,linestyle为网格线风式，-：实线，--：虚线
#设置刻度字体风格
plt.xticks(rotation=0,fontsize=12) #rotation表示刻度字体旋转的方向
plt.yticks(rotation=0,fontsize=12)
#设置y轴的范围
plt.ylim(0,30)
plt.plot([0,10],[0,20],color='red',linewidth=2,linestyle='-')   #坐标为（0，0）、（10，20）：[]第一个里的都为x的坐标，[]第二个为y的坐标
#自动优化排版
plt.tight_layout()
#显示图表
plt.show()
```
