```python
import matplotlib.pyplot as plt
from matplotlib import rcParams #jupyter不带字体(1月···)，要设置
rcParams['font.family']='STHeiti' #mac的   window:SimHei
#创建图表
plt.figure(figsize=(10,5)) #图表纸张的大小
#要绘图的数据
subjects=['语文','数学','英语','科学']
scores=[85,93,60,88]
#绘制柱状图
plt.bar(subjects,scores,
        label='小红',    
        color='orange',   
        linewidth=0.6,     
        )        
#添加标题
plt.title('2025年成绩情况',color='red',fontsize=20)
#添加坐标轴的标签
plt.xlabel('科目',fontsize=10)
plt.ylabel('分数（满分100）',fontsize=10)
#添加图例的位置
plt.legend(loc='upper left')
#添加网格线
plt.grid(axis='y',alpha=0.1,color='blue',linestyle='--')
#设置刻度字体风格
plt.xticks(rotation=0,fontsize=12) #rotation表示刻度字体旋转的方向
plt.yticks(rotation=0,fontsize=12)
#设置y轴的范围
plt.ylim(0,100)
#每个数据点显示数值
for x,y in zip(subjects,scores):
    #将打包遍历出的放在相对应地x,y轴上，水平剧中，线的上面
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10)
#自动优化排版
plt.tight_layout()
#显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams #jupyter不带字体(1月···)，要设置
rcParams['font.family']='STHeiti' #mac的   window:SimHei
#创建图表
plt.figure(figsize=(10,5)) #图表纸张的大小
#要绘图的数据
contries=['united States','China','Japan','Germany','India']
gdp=[92,83,60,32,8]
#绘制条形图
plt.barh(contries,gdp,    
        color='orange',      
        )        
#添加标题
plt.title('2025年GDP排名',color='red',fontsize=20)
#添加坐标轴的标签
plt.xlabel('GDP',fontsize=10)
plt.ylabel('国家',fontsize=10)
#自动优化排版
plt.tight_layout()
#显示图表
plt.show()
```
