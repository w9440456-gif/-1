```python
#绘制饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams #jupyter不带字体(1月···)，要设置
rcParams['font.family']='STHeiti' #mac的   window:SimHei
#创建图表
plt.figure(figsize=(10,5)) #图表纸张的大小
#要绘图的数据
things=['学习','娱乐','运动','睡觉','其他']
times=[6,4,1,8,5]
colors=['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499']
#绘制饼图图
plt.pie(times,labels=things,
       autopct='%.1f%%',   #保留一位浮点数,最后两个%%，显示一个%
       startangle=90,      #起始的角度，学习是起始在这里
       colors=colors)
#添加标题
plt.title('一天的时间分布',color='red',fontsize=20)
#自动优化排版
plt.tight_layout()
#显示图表
plt.show()
```

```python
#绘制环形图：在饼图中间挖了个圈
import matplotlib.pyplot as plt
from matplotlib import rcParams #jupyter不带字体(1月···)，要设置
rcParams['font.family']='STHeiti' #mac的   window:SimHei
#创建图表
plt.figure(figsize=(10,5)) #图表纸张的大小
#要绘图的数据
things=['学习','娱乐','运动','睡觉','其他']
times=[6,4,1,8,5]
colors=['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499']
#绘制环形图
plt.pie(times,labels=things,
       autopct='%.1f%%',   #保留一位浮点数,最后两个%%，显示一个%
       startangle=90,      #起始的角度，学习是起始在这里
       colors=colors,
       wedgeprops={'width':0.5},  #挖的圆心半径所占的比例
       pctdistance=0.8)      #百分比的位置
#添加标题
plt.title('一天的时间分布',color='red',fontsize=20)
plt.text(0,0,'总计：\n\n100%',ha='center',va='bottom',fontsize=10) #圈的内容
#自动优化排版
plt.tight_layout()
#显示图表
plt.show()
```

```python
#绘制爆炸式饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams #jupyter不带字体(1月···)，要设置
rcParams['font.family']='STHeiti' #mac的   window:SimHei
#创建图表
plt.figure(figsize=(10,5)) #图表纸张的大小
#要绘图的数据
things=['学习','娱乐','运动','睡觉','其他']
times=[6,4,1,8,5]
colors=['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499']
explode=[0.1,0,0,0,0]
#绘制爆炸式饼图
plt.pie(times,labels=things,
       autopct='%.1f%%',   #保留一位浮点数,最后两个%%，显示一个%
       startangle=0,      #起始的角度，学习是起始在这里
       colors=colors,
       explode=explode,  #设置突出块
       shadow=True)      #增加阴影效果
#添加标题
plt.title('一天的时间分布',color='red',fontsize=20)
#自动优化排版
plt.tight_layout()
#显示图表
plt.show()
```
