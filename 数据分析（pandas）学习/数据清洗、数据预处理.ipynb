```python
#缺失值的处理
#nan:not a number
import pandas as pd
import numpy as np
s=pd.Series([12,25,np.nan,None,pd.NA])
print(s)
df=pd.DataFrame([[1,2,np.nan],[2,3,5],[None,4,6]],columns=['第1列','第2列','第3列'])
#查看是否是缺失值
print(s.isna())
print(s.isnull())
print(df.isna())
print(df.isnull())
```

```python
print(df.isna().sum()) #每一列的缺失值个数，如果要看行，在sum()里加axis=1
print(s.isna().sum())
```

```python
#剔除缺失值
print(s)
print(s.dropna()) 
print('\n')
print(df)
print(df.dropna())  #删除一整行的
print(df.dropna(how='all')) #只有整行都是缺失值才删一行
print(df.dropna(thresh=1)) #每行只要有1个非缺失值，保留
print(df.dropna(axis=1)) #删除一整列的（axis=0沿上下操作，axis=1沿左右）
print(df.dropna(subset=['第1列'])) #如果第一列有缺失值，则删除该缺失值的行
```

```python
#填充缺失值
df=pd.read_csv('data/weather_withna.csv')
df.isna().sum(axis=0) #每1列的缺失值个数 
print(df.fillna({'temp_max':20,'wind':2.5}).tail()) #使用字典填充固定值，取最后5行出来
print(df.fillna(df[['temp_max','wind']].mean().tail())) #使用统计值填充，取最后5行
print(df.ffill().tail()) #用相邻前面的值填充
print(df.bfill().tail()) #用相邻后面的值填充
```

```python
#重复数据的处理
import pandas as pd
data={
    'name':['alice','alice','bob','alice','jack','bob'],
    'age':[26,25,30,25,35,30],
    'city':['NY','NY','LA','NY','SF','LA']
}
df=pd.DataFrame(data)
df
```

```python
df.duplicated() #一整条记录都是一样的，标记为重复，返回ture
df.drop_duplicates(subset=['name']) #根据指定列里是否重复，去除重复后的一行
df.drop_duplicates(subset=['name'],keep='last') #去除重复前的一行，保留最新的
```

```python
#数据变形
import pandas as pd
data={
    'id':[1,2],
    'name':['alice','bob'],
    'math':[90,85],
    'english':[88,92],
    'science':[95,89]
}
df=pd.DataFrame(data)
print(df)
df.T #行列转置
#宽表变长表
df2=pd.melt(df,id_vars=['id','name'],var_name='科目',value_name='分数')
df2.sort_values('name') #按姓名排序
print(df2)
#长表转宽表
pd.pivot(df2,index=['id','name'],columns='科目',values='分数')
```

```python
#分列
data={
    'id':[1,2],
    'name':['alice smith','bob smith'],
    'math':[90,85],
    'english':[88,92],
    'science':[95,89]
}
df=pd.DataFrame(data)
df[['first','last']]=df['name'].str.split(' ',expand=True) #用空格来字符串分列，
df=pd.read_csv('data/sleep.csv')
df=df[['person_id','blood_pressure']] #取两列出来
df[['high','low']]=df['blood_pressure'].str.split('/',expand=True) #分完以后是obejct,要转化为其他，方便后续统计
df['high']=df['high'].astype('int64')
df['high']=df['high'].astype('int64')
df.high.mean()
df.low.mean()
```

```python
#数据分箱(pd.cut):本质就是分区间段
data={
    'employee_id':[100,101,102,103,104,105,106,107,108,109],
    'salary':[24000,17000,17000,9000,6000,4800,4800,4200,12000,9000]
}
df=pd.DataFrame(data)
pd.cut(df['salary'],bins=2) #bins=n,分为n段区间，起始值和终点值是数据的最小值和最大值
pd.cut(df['salary'],bins=2).value_counts()
```

```python
pd.cut(df['salary'],bins=[0,10000,20000,30000]) #按列表分为0-10000，10000-20000
pd.cut(df['salary'],bins=[0,10000,20000,30000]).value_counts()
df['收入范围']=pd.cut(df['salary'],bins=[0,10000,20000,30000],labels=['低','中','高'])
df
```

```python
pd.qcut(df['salary'],3) #分成3段区间，每段的统计次数近似平均
pd.qcut(df['salary'],3).value_counts()
```
