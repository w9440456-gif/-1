# 电商页面A/B测试  
## 背景  
   某电商公司非常注重自己的**落地页**设计，希望通过改进设计来提高**转化率**。以往该公司全年转化率平均在13%左右，现在希望设计的新页面能够带来更高的转化率，希望新页面的转化率能有2%的提升，达到15%。在正式推出新页面之前，该公司希望通过AB测试在小范围的用户中进行测试，以确保新页面的效果能够达到预期目标。  
## ABtest  
通常包括以下流程：  
### 1. 实验改动点确定  
在ABtest开始前，需要和业务部门确定准确的实验改动点，即实验变量。本项目中实验变量为落地页的不同设计风格。  

### 2. 确定实验目标和衡量指标  
本实验的目标为通过AB测试确定新的页面设计是否能够提高2%的转化率，衡量指标为转化率。  

### 3.提出假设  
   假设是对于某个特定变化我们所期望的结果，也是后续实验的基础，我们需要在后续实验中通过数据验证这个假设是否成立。如果验证成立，我们可以将这个变化推广到全部用户。如果验证不成立，则需要继续优化这个假设或者放弃这个修改方案，以寻找更好的变化。  
  在这个实验中，我们希望新页面可以提升2%的转化率，原则上我们应选择右侧单尾检验，因为我们的假设是新页面的转化率要大于旧页面的转化率。但是，在本案例中，我们并不能确定新页面的性能一定比当前的页面更好，转化率更高只是我们的希望。我们要先验证它们存在差异，所以，这里选择双尾检验。  
  如果实验的目的是验证新设计的落地页转化率是否明显高于旧设计，就可以选择单尾检验。但如果不确定两种设计是否存在差异，则应选择双尾检验。  
			
  通常，将我们期望达到的结果放到备择假设上（拒绝原假设，接受备择假设更有说服力），在本实验中，提出假设如下：  
				H0：P0=P1  
				H1：P0≠P1  
  其中，P0表示旧版页面设计的转化率，P1表示新版页面设计的转化率  

### 4.实验分组（流量分割）  
  这一步目的是确保实验组和对照组的样本分布均匀且具有随机性，从而最大程度地减少其他因素对实验结果的干扰。具体来说，合理的样本分割应遵循随机化原则，使实验组和对照组在性别、年龄、地区、设备类型、活跃程度等关键特征上尽可能一致，以避免潜在的系统性偏差。通过这样的方法，可以确保在实验期间，只有实验改动点这一因素对核心指标产生影响，而其他无关变量被有效控制。  
	此外，合理的随机分组有助于避免辛普森悖论的发生。辛普森悖论指的是，当数据被分组分析时，总体趋势可能会被分组内的特定趋势掩盖或逆转，从而得出错误的结论。这种现象往往是由于样本分布不均匀导致的。  
		
  本次实验中，分组 如下：  
  * 对照组：用户看到旧版的落地页。  
  *   实验组：用户看到新版的落地页。  
  
  为了便于计算转化率，对用户购买行为进行记录：  
	         	0：代表用户在测试期间没有购买产品  
				    1：代表用户在测试期间购买了产品  
  后续通过求平均值，得到每个组的转化率。  
				
### 5. 计算实验最小样本量和实验周期  
* **最小样本量**  
AB测试是用小样本来推断总体，总体的分布有多种可能 ，可能是正态分布，也可能是偏态分布。  
根据中心极限定理：在样本量足够的的情况下（>30）,样本均值的分布，趋近于正态分布，正态分布曲线由均值和方差决定，该分布均值的均值E( ̅x)趋近于总体均值u。  
可以通过抽样，借助正态分布来估计置信区间，实现参数检验（t检验等）。  

样本量越大，估计就越准确，但成本也就越高。如何选定合适的最小的样本量？  
最小样本量计算公式如下：  
													![Image Name](https://cdn.kesci.com/upload/sy1gvj7hu6.png?imageView2/0/w/960/h/960)  
						
n为对照组和实验组各组所需要的样本量，总样本即为2n  
σ为标准差  
Δ为效应大小，指预期的两组差异幅度，在本项目中，Δ等于实验组的转化率减去对照组的转化率  
α为犯第一类错误的概率，即弃真错误的概率。  
𝛼=𝑃(拒绝 𝐻0∣𝐻0 为真)  
β为犯第二类错误的概率，即纳伪错误的概率。  
β=P(接受 H0∣H0 为假)  
一般情况下我们设置：  
显著性水平：α=0.05（或0.01）  
统计功效（1−β）：β=0.2，即指检测到真实差异的概率，通常设为 0.8。在此案例中就是，如果新版落地页真的比旧版转换率要高，该测试有80%的概率能检测出这个状况。  
当衡量指标为比率类指标时，标准差计算公式为：  
											![Image Name](https://cdn.kesci.com/upload/sy1huhxk4p.png?imageView2/0/w/960/h/960)  
其中，PA和PB分别是对照组和实验组的观测值。在此案例中PA=13%，PB=15%  
代入数据可求得，本实验 的最小 样本量为 N = 4716。  



```python
(0.13*(1-0.13)+0.15*(1-0.15))/(0.15-0.13)**2*(1.96+0.84)**2
```

**使用python完成最小样本量计算：**

```python
import numpy as np 
import pandas as pd 
import scipy.stats as stats 
import statsmodels.stats.api as sms 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#计算效应大小，该函数用于比较两个比例之间的标准化差异
effect_size = sms.proportion_effectsize(0.13, 0.15)
```

```python
effect_size
```

```python
#基于效应大小、功效、显著性水平等参数，计算每组所需的样本量
required_n = sms.NormalIndPower().solve_power( 
    effect_size, 
    power=0.8, #统计功效
    alpha=0.05, #显著水平
    ratio=1 ,
    )
```

```python
required_n
```

```python
np.ceil(required_n) #向上取整
```

### 6. 进行灰度测试  
在正式开启ABtest之前，为了确保实验的改动不会对系统或用户体验造成巨大的异常影响，通常需要先选取一小部分用户样本进行灰度测试，旨在通过对少量用户的观察，评估实验改动点的风险和稳定性。  
通过灰度测试，可以验证实验的功能是否正常，用户体验是否符合预期，以及核心指标是否出现重大波动。如果灰度测试结果良好且风险可控，便可逐步扩大分流比例，最终进入正式的ABtest阶段；否则，需对实验方案或改动点进行调整，避免全面上线后带来不可控的负面影响。

接下来需要进行假设检验来进行客观的分析。

####  假设检验  
这里是对新旧版页面的转化率是否相等进行检验，属于对比例进行检验。样本量足够大，总体方差未知，所以采用Z检验的方法，构造Z统计量：  
Z=（p1-p2）/sqrt[p1×(1-p1)/n1+p2×(1-p2)/n2]

利用Python中的statsmodels.stats.proportion模块可以来计算P值和置信区间：

```python
from statsmodels.stats.proportion import proportions_ztest,proportion_confint
# proportions_ztest需要传入每一组的数量和成功转化的数量
# 取出converted值
control_results = test_data[test_data["group"] == 'control']["converted"]
treatment_results = test_data[test_data["group"] == 'treatment']["converted"]
```

```python
n_con = control_results.count()
n_treat = treatment_results.count()
nobs = [n_con,n_treat]
successes = [control_results.sum(),treatment_results.sum()]
```

```python
nobs
```

```python
successes
```

```python
# 计算Z值和P值
z_stat,p_val = proportions_ztest(successes,nobs)

# 计算临界值
critical_value = stats.norm.ppf(1-0.025)

# 计算对照组和实验组95%置信区间
(lower_con,lower_treat),(upper_con,upper_treat) = proportion_confint(successes,nobs,alpha=0.05)
```

```python
# 打印结果
print(f"p_value : {p_val:.3f}")
print(f"z_statistic : {z_stat:.2f}")
print(f"临界值为 : {critical_value:.2f}")
print("落入拒绝域拒绝原假设,两版本页面存在差异" if abs(z_stat) > critical_value else "接受原假设，两版本页面不存在差异") 
print(f"ci 95% for control group : [{lower_con:.3f},{upper_con:.3f}]")
print(f"ci 95% for control group : [{lower_treat:.3f},{upper_treat:.3f}]")
```

## 分析结果与建议  


从上面的输出结果可以看出，p值远高于显著水平0.05，并且z统计量的值也是在接受域，也就是说新旧两版页面的转化率并没有显著的不同。  

此外，从置信区间来看，相比于旧版页面，新版页面转化率的置信区间  
* 有一定的增加，但是差异非常微小  
* 上限接近13%，明显低于15%  

综上，新版页面的真实转化率更可能接近基线值13%，与旧版页面并无明显差异，说明当前的修改并不有效 ，下一步应该继续对页面设计进行其他方面的优化，也可以通过用户调研的方式探索更能满足用户需求的页面设计。
