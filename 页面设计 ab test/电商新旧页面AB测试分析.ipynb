```python
import numpy as np 
import pandas as pd 
import scipy.stats as stats 
import statsmodels.stats.api as sms 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#è®¡ç®—æ•ˆåº”å¤§å°ï¼Œè¯¥å‡½æ•°ç”¨äºæ¯”è¾ƒä¸¤ä¸ªæ¯”ä¾‹ä¹‹é—´çš„æ ‡å‡†åŒ–å·®å¼‚
effect_size = sms.proportion_effectsize(0.13, 0.15)
```

```python
effect_size
```

```python
#åŸºäºæ•ˆåº”å¤§å°ã€åŠŸæ•ˆã€æ˜¾è‘—æ€§æ°´å¹³ç­‰å‚æ•°ï¼Œè®¡ç®—æ¯ç»„æ‰€éœ€çš„æ ·æœ¬é‡
required_n = sms.NormalIndPower().solve_power( 
    effect_size, 
    power=0.8, #ç»Ÿè®¡åŠŸæ•ˆ
    alpha=0.05, #æ˜¾è‘—æ°´å¹³
    ratio=1 ,
    )
```

å¯è§ï¼Œæ–°æ—§ä¸¤ç‰ˆè½åœ°é¡µçš„è½¬åŒ–ç‡éå¸¸æ¥è¿‘ï¼Œç›¸å¯¹æ—§ç‰ˆï¼Œæ–°ç‰ˆçš„è½¬åŒ–ç‡é«˜äº†0.8%  
é‚£ä¹ˆæˆ‘ä»¬å¯ä»¥è¯´æ–°ç‰ˆè½åœ°é¡µçš„è½¬åŒ–ç‡æ›´å¥½å—ï¼Ÿè¿™æ ·çš„å·®å¼‚åœ¨ç»Ÿè®¡å­¦ä¸Šæ˜¯æ˜¾è‘—çš„å—ï¼Ÿè¿˜åªæ˜¯å¶ç„¶çš„ç»“æœï¼Ÿ  
æ¥ä¸‹æ¥éœ€è¦è¿›è¡Œå‡è®¾æ£€éªŒæ¥è¿›è¡Œå®¢è§‚çš„åˆ†æã€‚

####  å‡è®¾æ£€éªŒ  
è¿™é‡Œæ˜¯å¯¹æ–°æ—§ç‰ˆé¡µé¢çš„è½¬åŒ–ç‡æ˜¯å¦ç›¸ç­‰è¿›è¡Œæ£€éªŒï¼Œå±äºå¯¹æ¯”ä¾‹è¿›è¡Œæ£€éªŒã€‚æ ·æœ¬é‡è¶³å¤Ÿå¤§ï¼Œæ€»ä½“æ–¹å·®æœªçŸ¥ï¼Œæ‰€ä»¥é‡‡ç”¨Zæ£€éªŒçš„æ–¹æ³•ï¼Œæ„é€ Zç»Ÿè®¡é‡ï¼š  
Z=ï¼ˆp1-p2ï¼‰/sqrt[p1Ã—(1-p1)/n1+p2Ã—(1-p2)/n2]

åˆ©ç”¨Pythonä¸­çš„statsmodels.stats.proportionæ¨¡å—å¯ä»¥æ¥è®¡ç®—På€¼å’Œç½®ä¿¡åŒºé—´ï¼š

```python
from statsmodels.stats.proportion import proportions_ztest,proportion_confint
# proportions_ztestéœ€è¦ä¼ å…¥æ¯ä¸€ç»„çš„æ•°é‡å’ŒæˆåŠŸè½¬åŒ–çš„æ•°é‡
# å–å‡ºconvertedå€¼
control_results = test_data[test_data["group"] == 'control']["converted"]
treatment_results = test_data[test_data["group"] == 'treatment']["converted"]
```

```python
n_con = control_results.count()
n_treat = treatment_results.count()
nobs = [n_con,n_treat]
successes = [control_results.sum(),treatment_results.sum()]
```

```python
nobs
```

```python
successes
```

```python
# è®¡ç®—Zå€¼å’ŒPå€¼
z_stat,p_val = proportions_ztest(successes,nobs)

# è®¡ç®—ä¸´ç•Œå€¼
critical_value = stats.norm.ppf(1-0.025)

# è®¡ç®—å¯¹ç…§ç»„å’Œå®éªŒç»„95%ç½®ä¿¡åŒºé—´
(lower_con,lower_treat),(upper_con,upper_treat) = proportion_confint(successes,nobs,alpha=0.05)
```

```python
# æ‰“å°ç»“æœ
print(f"p_value : {p_val:.3f}")
print(f"z_statistic : {z_stat:.2f}")
print(f"ä¸´ç•Œå€¼ä¸º : {critical_value:.2f}")
print("è½å…¥æ‹’ç»åŸŸæ‹’ç»åŸå‡è®¾,ä¸¤ç‰ˆæœ¬é¡µé¢å­˜åœ¨å·®å¼‚" if abs(z_stat) > critical_value else "æ¥å—åŸå‡è®¾ï¼Œä¸¤ç‰ˆæœ¬é¡µé¢ä¸å­˜åœ¨å·®å¼‚") 
print(f"ci 95% for control group : [{lower_con:.3f},{upper_con:.3f}]")
print(f"ci 95% for control group : [{lower_treat:.3f},{upper_treat:.3f}]")
```
